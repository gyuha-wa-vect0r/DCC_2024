```python
import torch
import os
import cv2
import numpy as np
```

```python
# YOLOv5 ëª¨ë¸ ë¡œë“œ (ì‚¬ì „ í•™ìŠµëœ ëª¨ë¸ ì‚¬ìš©)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # 'yolov5s' ë“± ì›í•˜ëŠ” ëª¨ë¸ ì„ íƒ ê°€ëŠ¥
```

```python
# ìž…ë ¥ ë° ì¶œë ¥ ê²½ë¡œ ì„¤ì •
input_dirs = [
    '/home/gyuha_lee/DCC2024/dataset/training_image',
]
output_dirs = [
    '/home/gyuha_lee/DCC2024/dataset/bg_remove/yolov5/training_image_no_bg',
]
```

```python
# ì¶œë ¥ ë””ë ‰í† ë¦¬ ìƒì„±
for output_dir in output_dirs:
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
```

```python
# ëª¨ë“  ì´ë¯¸ì§€ì— ëŒ€í•´ YOLO ëª¨ë¸ ì ìš© ë° ë§ˆìŠ¤í‚¹
for input_dir, output_dir in zip(input_dirs, output_dirs):
    i = 1
    for img_file in os.listdir(input_dir):
        if img_file.endswith('.jpg'):
            input_path = os.path.join(input_dir, img_file)
            output_path = os.path.join(output_dir, img_file)

            # ì´ë¯¸ì§€ ë¶ˆëŸ¬ì˜¤ê¸°
            image = cv2.imread(input_path)

            # ëª¨ë¸ ì¶”ë¡ 
            results = model(input_path)

            # íƒì§€ ê²°ê³¼ì—ì„œ ë§ˆìŠ¤í¬ ìƒì„±
            detected = results.pandas().xyxy[0]
            mask = np.zeros(image.shape[:2], dtype=np.uint8)  # ë¹ˆ ë§ˆìŠ¤í¬ ìƒì„±

            for index, row in detected.iterrows():
                # ì˜·ê³¼ ê´€ë ¨ëœ ê°ì²´ë§Œ í•„í„°ë§ (í•„ìš”ì— ë”°ë¼ ì¶”ê°€)
                if row['name'] in ['person', 'shirt', 'pants', 'dress']:
                    x_min, y_min, x_max, y_max = int(row['xmin']), int(row['ymin']), int(row['xmax']), int(row['ymax'])
                    # íƒì§€ëœ ê°ì²´ ì˜ì—­ì— ëŒ€í•´ ë§ˆìŠ¤í¬ ì„¤ì •
                    mask[y_min:y_max, x_min:x_max] = 255

            # ì›ë³¸ ì´ë¯¸ì§€ì— ë§ˆìŠ¤í¬ ì ìš©
            masked_image = cv2.bitwise_and(image, image, mask=mask)

            # ì•ŒíŒŒ ì±„ë„ ì¶”ê°€ (ë°°ê²½ì„ íˆ¬ëª…í•˜ê²Œ ì²˜ë¦¬)
            b, g, r = cv2.split(masked_image)
            alpha = mask  # ë§ˆìŠ¤í¬ë¥¼ ì•ŒíŒŒ ì±„ë„ë¡œ ì‚¬ìš©
            masked_image_rgba = cv2.merge([b, g, r, alpha])

            # ê²°ê³¼ ì €ìž¥
            output_path = output_path.replace('.jpg', '.png')  # ì•ŒíŒŒ ì±„ë„ì„ í¬í•¨í•˜ë¯€ë¡œ .pngë¡œ ì €ìž¥
            cv2.imwrite(output_path, masked_image_rgba)
            print(i, img_file, end='\n')
            i += 1
```
